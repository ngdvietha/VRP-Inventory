```python
import gurobipy as gp
from gurobipy import GRB
```

```python
import pandas as pd
import numpy as np
```

```python
#Import những tham số đầu vào của các node

#Import những biến đầu vào về tồn kho

minimumStock = 0
maximumStock = 10000000
# maximumStock1 = 300
# maximumStock2 = 300
stockCost = 0.75

#Khai báo số ngày tiến hành chạy mô hình
Time = list(range(1, 4))

#Insert dữ liệu của nhà sản xuất (NodeID, X, Y, ProductionRate, FirstInventory, NodeType, MinimumOrder)
Suppliers = pd.read_excel("Input data.xlsx", sheet_name='Supplier').set_index('NodeID').to_dict(orient = 'index')

#Insert dữ liệu của Distributor ((NodeID, X, Y, FirstInventory, NodeType, StartTime, EndTime)
Distributor = pd.read_excel("Input data.xlsx", sheet_name='Distributor').set_index('NodeID').to_dict(orient = 'index')

#Insert dữ liệu của Retailer (NodeID, X, Y, NodeType, StartTime, EndTime, Demand)
Retailer = pd.read_excel("Input data.xlsx", sheet_name='Retailer').set_index('NodeID').to_dict(orient = 'index')   

#Đổi lại kiểu dữ liệu của Demand về thành list
for r, d in Retailer.items():
     d['Demand'] = d['Demand'].split(',')

#Thêm một dictionary tự gen ra distributor return
returnDistributor = pd.read_excel("Input data.xlsx", sheet_name='returnDistributor').set_index('NodeID').to_dict(orient = 'index')
```

```python
list(Distributor.keys())
```

```python
#Import dữ liệu những tham số đầu khác

#Tính chất xe trong echelon1 (xe ở echelon 1 không có starttime và endtime)
VehicleE1 = pd.read_excel("Input data.xlsx", sheet_name='VehicleE1').set_index('VehicleID').to_dict(orient = 'index')

#Tính chất xe trong echelon2 (xe ở echelon 1 không có starttime và endtime)
VehicleE2 = pd.read_excel("Input data.xlsx", sheet_name='VehicleE2').set_index('VehicleID').to_dict(orient = 'index')
```

```python
#Tập hợp tất cả các điểm của echelon 1
N1 = {**Suppliers, **Distributor}

#Tập hợp tất cả các điểm của echelon 2
N2 = {**Retailer,**Distributor, **returnDistributor}
```

```python
N2
```

```python
#Cung đường của echelon 1, các nhà cung cấp sẽ không đi qua lại lẫn nhau, các distributor không qua lại lẫn nhau
E1 = [] 

for i in N1.keys():
    for j in N1.keys():
                if i != j and N1[i]['NodeType'] != N1[j]['NodeType']:
                    E1.append((i, j))


#Cung đường của echelon 2, các retailer có thể qua lại lẫn nhau, các distributor không qua lại lẫn nhau
E2 = [] 

for k1, v1 in N2.items():
    for k2, v2 in N2.items():
                if k1 != k2 and \
                    (v1['NodeType'] != 'Distributor' or v2['NodeType'] != 'Distributor') and \
                    (k1 not in returnDistributor.keys()) and \
                    (v1['NodeType'] != 'Distributor' or v2['NodeType'] != 'returnDistributor') and \
                    (v1['NodeType'] != 'Retailer' or v2['NodeType'] != 'Distributor'):
                        E2.append((k1, k2))

E2
```

```python
#Notes sau này sẽ cần phải API để lấy được quãng đường thực tế ở echelon 2

#Khai báo thông tin về quãng đường của echelon 2
def euclidean_distance(x1, x2, x3, x4):
    return np.sqrt((x1 - x2) ** 2 + (x3 - x4) ** 2)

distance_list = {}

for e1 in E1:
    distance_list[e1] = euclidean_distance(N1[e1[0]]['X'], N1[e1[1]]['X'], N1[e1[0]]['Y'], N1[e1[1]]['Y'])
    
for e2 in E2:
    distance_list[e2] = euclidean_distance(N2[e2[0]]['X'], N2[e2[1]]['X'], N2[e2[0]]['Y'], N2[e2[1]]['Y'])

distance_list[('R1', 'R2')]
```

```python
# Create a new model
model = gp.Model("VRP")

# Create variables
#KHAI BÁO ECHELON 1
#Biến quyết định giao hàng từ manufacturer m tới distributor d bằng xe v tại thời điểm t trong E1
u = {(e1, VE1, T): model.addVar(vtype=GRB.BINARY, name='u_{}_{}_{}'.format(e1, VE1, T))
     for e1 in E1
     for VE1 in VehicleE1.keys()
     for T in Time}

#Biến lượng giao hàng từ manufacturer m tới distributor d bằng xe v tại thời điểm t trong E1
delivery = {(e1, VE1, T): model.addVar(vtype=GRB.INTEGER, lb=0, name='delivery_{}_{}_{}'.format(e1, VE1, T))
     for e1 in E1
     for VE1 in VehicleE1.keys()
     for T in Time}

#Inventory level của manufacturer m
inventoryLevelM = {(M, T): model.addVar(vtype=GRB.INTEGER, lb=0, name='inventoryLevelM_{}_{}'.format(M, T))
     for M in Suppliers.keys()
     for T in Time}

#Inventory level của distributor d
inventoryLevelD = {(D, T): model.addVar(vtype=GRB.INTEGER, lb=0, name='inventoryLevelD_{}_{}'.format(D, T))
     for D in Distributor.keys()
     for T in Time}

#KHAI BÁO ECHELON 2
#Biến quyết định giao hàng từ distributor d tới retailer r bằng xe v tại thời điểm t trong E2
x = {(e2, VE2, T): model.addVar(vtype=GRB.BINARY, name='x_{}_{}_{}'.format(e2, VE2, T))
     for e2 in E2
     for VE2 in VehicleE2.keys()
     for T in Time}

#Biến quyết định thời gian giao hàng từ distributor d tới retailer r bằng xe v tại thời điểm t trong E2
s = {(n2, VE2, T): model.addVar(vtype=GRB.CONTINUOUS, name='s_{}_{}_{}'.format(n2, VE2, T))
     for n2 in N2.keys()
     for VE2 in VehicleE2.keys()
     for T in Time}
```

```python
#Khai báo các ràng buộc

#Echelon 1
#(1) Ràng buộc các xe xuất phát từ supplier 
for k1 in VehicleE1:
    for t in Time:
        model.addConstr(gp.quicksum(u[e1, k1, t] for e1 in E1 if e1[0] in Suppliers) <= 1, f'visitOnceE1_{k1}_{t}')
        
#(2) Ràng buộc xe đi từ supplier nào thì về supplier tương ứng
for k1 in VehicleE1:
    for t in Time:
        for e1 in E1:
            model.addConstr(u[e1, k1, t] == u[(e1[1],e1[0]), k1, t], f'mustReturnE1_{e1}_{k1}_{t}')
            
#(3) Lượng vận chuyển nhỏ hơn số lượng hiện có của supplier
for su in Suppliers:
    for t in Time:
        model.addConstr(gp.quicksum(delivery[e1, k1, t] for e1 in E1 for k1 in VehicleE1 if e1[0] == su) <= inventoryLevelM[su,t],f'stockE1_{su}_{t}')

#(4) Ràng buộc về mặt logic
for e1 in E1:
    for k1 in VehicleE1:
        for t in Time:
            model.addConstr(10000000*u[e1,k1,t] >= delivery[e1,k1,t],f'existDelivery_{e1}_{k1}_{t}')

#(5) Ràng buộc lượng vận chuyển thoả mãn tải trọng xe trong E1
for e1 in E1:
    for k1 in VehicleE1:
        for t in Time:
            model.addConstr(delivery[e1,k1,t] <= VehicleE1[k1]['Capacity'],f'capLimitE1_{e1}_{k1}_{t}')
            
#(6) Lượng delivery phải lớn hơn lượng minimum được quy định bởi supplier
for e1 in E1:
    for k1 in VehicleE1:
        for t in Time:
            for su in Suppliers:
                if e1[0] == su:
                    model.addConstr(u[e1,k1,t]*Suppliers[su]['MinOrder'] <= delivery[e1,k1,t],f'minimumOrder_{e1}_{k1}_{t}')
    
#Nối Echelon 1 và Echelon 2
#(7) Khởi tạo inventorylevelM ngày 1 cho supplier m
for su in Suppliers:
    for t in Time:
        if t==1:
            model.addConstr(inventoryLevelM[su,t] == Suppliers[su]['FirstInventory']
    + Suppliers[su]['ProductionRate']
    - gp.quicksum(delivery[e1,k1,t] for e1 in E1 for k1 in VehicleE1 if e1[0] == su),f'firstInventoryM_{su}_{t}')           

#(8) Update inventoryLevelM các ngày tiếp theo cho supplier m
        if t>1:
            model.addConstr(inventoryLevelM[su,t] == inventoryLevelM[su,t-1]
    + Suppliers[su]['ProductionRate']
    - gp.quicksum(delivery[e1,k1,t] for e1 in E1 for k1 in VehicleE1 if e1[0] == su),f'updateInventoryM_{su}_{t}')

#(9) Khởi tạo inventoryLevelD ngày 1 cho distributor d
for d in Distributor:
    for t in Time:
        if t==1:
            model.addConstr(inventoryLevelD[d,t] == Distributor[d]['FirstInventory']
    + gp.quicksum(delivery[e1,k1,t] for e1 in E1 for k1 in VehicleE1 if e1[1] == d)
    - gp.quicksum(x[e2,k2,t]*int(Retailer[r]['Demand'][t-1]) for e2 in E2 for k2 in VehicleE2 for r in Retailer if VehicleE2[k2]['StartAt']==d and e2[1] == r),f'firstInventoryD_{d}_{t}')           

#(10) Update inventoryLevelD các ngày tiếp theo cho distributor d
        if t>1:
            model.addConstr(inventoryLevelD[d,t] == inventoryLevelD[d,t-1]
    + gp.quicksum(delivery[e1,k1,t] for e1 in E1 for k1 in VehicleE1 if e1[1] == d)
    - gp.quicksum(x[e2,k2,t]*int(Retailer[r]['Demand'][t-1]) for e2 in E2 for k2 in VehicleE2 for r in Retailer if VehicleE2[k2]['StartAt']==d and e2[1] == r),f'updateInventoryD_{d}_{t}')           

#(11) Minimum stock cho distributor d
for d in Distributor:
    for t in Time:
        model.addConstr(inventoryLevelD[d,t] >= minimumStock,f'minStockDistributor_{d}_{t}')  

#(12) Maximum stock cho distributor d (giống nhau)
        model.addConstr(inventoryLevelD[d,t] <= maximumStock,f'maxStockDistributor_{d}_{t}')  

# #Maximum stock cho distributor d (khác nhau)
#         if d=='D1':
#             model.addConstr(inventoryLevelD[d,t] <= maximumStock1,f'maxStockDistributor1_{d}_{t}') 
#         if d=='D2':
#             model.addConstr(inventoryLevelD[d,t] <= maximumStock2,f'maxStockDistributor2_{d}_{t}') 
```

```python

#Echelon 2
#(13) Mỗi retailer được visit 1 lần
for r in Retailer:
    for t in Time:
        model.addConstr(gp.quicksum(x[e2,k2,t] for e2 in E2 for k2 in VehicleE2 if e2[0]==r)==1, f'visitOnceE2_{r}_{t}') 

#(14) Ràng buộc tải trọng xe của E2
for k2 in VehicleE2:
    for t in Time:
        model.addConstr(gp.quicksum(int(Retailer[r]['Demand'][t-1])* gp.quicksum(x[e2,k2,t] for e2 in E2 if e2[0]==r) for r in Retailer) <= int(VehicleE2[k2]['Capacity']), f'capLimitE2_{k2}_{t}') 

#(15) Ràng buộc xe không nhất thiết rời đi
for k2 in VehicleE2:
    for t in Time:
        model.addConstr(gp.quicksum(x[e2,k2,t] for e2 in E2 if VehicleE2[k2]['StartAt']==e2[0])<=1, f'mustLeaveDepot_{k2}_{t}') 

#(16) Ràng buộc xe không được xuất phát ở những điểm distributor không được chỉ định
for k2 in VehicleE2:
    for t in Time:
        model.addConstr(gp.quicksum(x[e2,k2,t] for e2 in E2 if VehicleE2[k2]['StartAt']!=e2[0] and e2[0] in list(Distributor.keys()))==0, f'mustNotLeaveDepot_{k2}_{t}') 

#(17) Ràng buộc về tiếp hàng xong thì phải rời đi
for r in Retailer:
    for k2 in VehicleE2:
        for t in Time:
            model.addConstr(gp.quicksum(x[e2,k2,t] for e2 in E2 if e2[1]==r) - gp.quicksum(x[e2,k2,t] for e2 in E2 if e2[0]==r)==0, f'mustLeaveCustomer_{r}_{k2}_{t}') 

#(18) Ràng buộc xe phải về distributor 
for k2 in VehicleE2:
    for t in Time:
        model.addConstr(gp.quicksum(x[e2,k2,t] for e2 in E2 if e2[1] in returnDistributor)<=1, f'mustReturnDepot_{k2}_{t}') 

#(19) Ràng buộc xe phải về đúng distributor đã xuất phát
        model.addConstr(gp.quicksum(x[e2,k2,t] for e2 in E2 if e2[1] in returnDistributor and e2[1]!= VehicleE2[k2]['EndAt'])==0, f'mustNotReturnDepot_{k2}_{t}')

#(20) Ràng buộc về thời gian
for e2 in E2:
    for k2 in VehicleE2:
        for t in Time:
            model.addConstr(x[e2,k2,t]*(s[e2[0],k2,t] + (distance_list[e2]/30) - s[e2[1],k2,t]) <=0, f'travelTime_{e2}_{k2}_{t}') 
    
#(21) Ràng buộc về limit của m (biến quyết định biến đổi tuyến tính)
# for e2 in E2:
#     for t in Time:
#         model.addConstr(m[e2, t] == N2[e2[0]]['EndTime'] + distance_list[e2]/30 - N2[e2[1]]['StartTime'], f'mLimit_{e2}_{t}') 

#(22) Hoạt động trong thời gian quy định của distributor/retailer trong Echelon 2
for n2 in N2:
    for k2 in VehicleE2:
        for t in Time:
            model.addConstr(N2[n2]['StartTime'] <= s[n2,k2,t], f'timeWindowE211_{k2}_{t}') 
            model.addConstr(N2[n2]['EndTime'] >= s[n2,k2,t], f'timeWindowE212_{k2}_{t}') 
            
#(23) Hoạt động trong thời gian quy định của xe 
            model.addConstr(VehicleE2[k2]['StartTime'] <= s[n2,k2,t], f'timeWindowE221_{k2}_{t}') 
            model.addConstr(VehicleE2[k2]['EndTime'] >= s[n2,k2,t], f'timeWindowE222_{k2}_{t}')
```

```python
#Khai báo hàm mục tiêu
#Chi phí E1
costE1 = gp.quicksum(u[e1, k1, t] * distance_list[e1] for e1 in E1 for k1 in VehicleE1 for t in Time)

#Chi phí E2
costE2 = gp.quicksum(x[e2, k2, t] * distance_list[e2] for e2 in E2 for k2 in VehicleE2 for t in Time)

#Chi phí inventory 
costInt = gp.quicksum(inventoryLevelD[d, t] * stockCost for d in Distributor for t in Time)

# Objective function: minimize costE1 + costE2 + costInt
model.setObjective(costE1 + 
    costE2 
    + costInt, GRB.MINIMIZE)
```

```python
# Time-limit
model.setParam('TimeLimit', 120)

# Optimize model
model.optimize()
```

```python
# Check if the optimization was successful
if model.status == GRB.OPTIMAL:
    print("Optimal solution found:\n")
    
    # Prepare data for DataFrame
    data = [v.VarName.split('_') + [v.X] for v in model.getVars()]
    
    # Create DataFrame
    frame = pd.DataFrame(data, columns=['var1', 'var2', 'var3', 'var4'])
    
    # Print the DataFrame
    print(frame)
    
    # Save to Excel
    frame.to_excel('SolutionX_gurobi.xlsx', index=False)

else:
    print("Optimization stopped with status:", model.status)
    # Prepare data for DataFrame
    data = [v.VarName.split('_') + [v.X] for v in model.getVars()]
    
    # Create DataFrame
    frame = pd.DataFrame(data, columns=['var1', 'var2', 'var3', 'var4', 'var5'])
    
    # Print the DataFrame
    print(frame)
    
    # Save to Excel
    frame.to_excel('SolutionX_gurobi.xlsx', index=False)
```
